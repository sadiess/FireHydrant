# Cut Pass Percentages

This is a short notebook used to generate the percentages for how many leptonJets passed each signal cut for each model.

Using "total" to mean total matched signal events that *have* passed the sorting cut:

```python
stats = {'current_vs_total': {'total': 358836, 'loose': 340606, 'medium': 294921, 'tight': 124132},
         'current_vs_elec': {'total': 36901, 'loose': 24115, 'medium': 3343, 'tight': 58},
         'current_vs_mu':{'total': 321788, 'loose': 316367, 'medium': 291479, 'tight': 124052},
         'electron_GM': {'total': 36901, 'loose': 30697, 'medium': 8618, 'tight': 444},
         'muon_GM': {'total': 321788, 'loose': 317673, 'medium': 295672, 'tight': 100859},
         'electron_neufrac': {'total': 37117, 'loose': 30771, 'medium': 8624, 'tight': 444},
         'muon_neufrac': {'total': 321719, 'loose': 318112, 'medium': 296008, 'tight': 100925},
         'electron_tracks': {'total': 41416, 'loose': 30801, 'medium': 8624, 'tight': 444},
         'muon_tracks': {'total': 317420, 'loose': 315852, 'medium': 294653, 'tight': 100633}
         }

for method in stats.keys():
    print("method: "+method)
    for cut in stats[method].keys():
        if cut != 'total':
            print("cut: "+cut)
            p = ((stats[method][cut])/(stats[method]['total']))*100
            print("pass %: ",p)
    print("")
```
